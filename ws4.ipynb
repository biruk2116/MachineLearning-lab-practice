```python


# 4.  Confusion Matrix, Accuracy, Precision, Recall, F1-Measure, ROC, AUC


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.metrics import roc_curve, auc, roc_auc_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder

data=pd.read_csv('dataset/predictions.csv')
actual=data['Actual']
predicted=data['Predicted']
# print (actual)
# print (predicted)

confusion_matrix = metrics.confusion_matrix(actual, predicted)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['Posetive', 'Negative'])
cm_display.plot()
plt.show()

precision = precision_score(actual, predicted, average='macro')
recall = recall_score(actual, predicted, average='macro')
f1 = f1_score(actual, predicted, average='macro')

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')



# Creating a instance of label Encoder.
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
data1 = le.fit_transform(data['Actual'])
data2 = le.fit_transform(data['Predicted'])

data3=[data1, data2]

fpr, tpr, thresholds = roc_curve(data1, data2)

print ("False Posetive rate = ", fpr)
print ("True Posetive Rate = ", tpr)
print ("The Threshold value = ", thresholds)

classes=['Posetive', 'Negative']
print(classification_report(actual, predicted, target_names=classes))

roc_auc = auc (fpr, tpr)
print(f'ROC AUC Score: {roc_auc:.2f}')

plt.figure(figsize=(4, 4))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

print(f"Calculated AUC: {roc_auc:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

data = pd.read_csv('dataset/iris.csv')

if len(data.columns) == 5:
    data.columns = ['SepLength', 'SepWidth', 'PetLength', 'PetWidth', 'Species']

data = data.replace(' ', np.nan)
data = data.replace('', np.nan)
data = data.replace('?', np.nan)

cols = ['SepLength', 'SepWidth', 'PetLength', 'PetWidth']
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')

data = data.dropna()

X = data.iloc[:, 0:4]
Y = data.iloc[:, 4]

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=100
)

clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()

print(export_text(clf, feature_names=list(X.columns)))

y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))

print(confusion_matrix(y_test, y_pred))

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1:", f1_score(y_test, y_pred, average='macro'))

sample = pd.DataFrame([[6.4, 1.8, 6.6, 2.1]], columns=X.columns)
print("Prediction:", clf.predict(sample))
```

```python
# 8. Decision tree using Gini index (CART) - FIXED VERSION

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('dataset/iris.csv')

# If no proper column names, assign them
if len(data.columns) == 5:
    data.columns = ['SepLength', 'SepWidth', 'PetLength', 'PetWidth', 'Species']

# Handle missing or invalid values
data = data.replace(r'^\s*$', np.nan, regex=True)
data = data.dropna()

# Convert numeric columns to float
for col in ['SepLength', 'SepWidth', 'PetLength', 'PetWidth']:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows where conversion resulted in NaN
data = data.dropna()

# Remove any leading/trailing spaces from Species names
data['Species'] = data['Species'].str.strip()

# Features and target
X = data.iloc[:, 0:4]
Y = data.iloc[:, 4]

print(f"Dataset shape: {data.shape}")
print(f"Unique classes in full dataset: {Y.unique()}")
print(f"Number of unique classes: {Y.nunique()}")

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=100
)

# Check classes in training and test sets
print(f"\nClasses in training set: {y_train.unique()}")
print(f"Classes in test set: {y_test.unique()}")

# Decision Tree (Gini)
clf = DecisionTreeClassifier(criterion="gini", random_state=100)
clf.fit(X_train, y_train)

# Prediction on test data
y_pred = clf.predict(X_test)

# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# Confusion Matrix - FIXED
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (without labels):")
print(conf_matrix)

# Get the actual classes that appear in the test set
actual_classes = np.unique(y_test)
print(f"\nActual classes in test set: {actual_classes}")

# Create confusion matrix with correct labels
conf_matrix_df = pd.DataFrame(
    conf_matrix, 
    index=actual_classes,  # Use only classes present in test set
    columns=actual_classes  # Use only classes present in test set
)

print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# Performance Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Predict new record
sample = pd.DataFrame(
    [[6.4, 1.8, 6.6, 2.1]],
    columns=X.columns
)

prediction = clf.predict(sample)
print("\nClass label for [6.4, 1.8, 6.6, 2.1]:", prediction[0])

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(
    clf,
    feature_names=X.columns,
    class_names=clf.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree using Gini Index (CART)")
plt.tight_layout()
plt.show()

# Tree Rules (Text)
tree_rules = export_text(clf, feature_names=list(X.columns))
print("\nDecision Tree Rules:\n")
print(tree_rules)
```

```python
# 10. Gaussian Naive Bayes on Iris dataset - FIXED VERSION

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Load the dataset
data = pd.read_csv('dataset/iris.csv')

# Assign proper column names
if len(data.columns) == 5:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
else:
    # If column names already exist, just ensure they're correct
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']

# Step 1a: Data cleaning - Handle missing or invalid values
print("Initial data shape:", data.shape)

# Replace empty strings or spaces with NaN
data = data.replace(r'^\s*$', np.nan, regex=True)

# Convert numeric columns to float
numeric_cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows with any NaN values
initial_rows = len(data)
data = data.dropna()
print(f"Dropped {initial_rows - len(data)} rows with invalid values")

# Clean the Class column (remove spaces, handle case)
data['Class'] = data['Class'].str.strip().str.lower()

print(f"Cleaned data shape: {data.shape}")
print(f"Unique classes: {data['Class'].unique()}")
print(f"Class distribution:\n{data['Class'].value_counts()}")

# Step 2: Separate features and target
X = data[numeric_cols]  # Use only numeric columns
y = data['Class']

# Verify data types
print(f"\nFeature data types:\n{X.dtypes}")
print(f"\nFirst few rows of features:\n{X.head()}")

# Step 3: Split dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nTraining set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
print(f"Classes in training: {y_train.unique()}")
print(f"Classes in test: {y_test.unique()}")

# Step 4: Initialize and train the Gaussian Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)

print("\nModel trained successfully!")

# Step 5: Predict class labels on the testing dataset
y_pred = gnb.predict(X_test)
print("\nFirst 10 predicted class labels for the test set:")
print(y_pred[:10])

# Step 6: Predict probabilities of each class for the testing dataset
pred_probabilities = gnb.predict_proba(X_test)
print("\nPredicted probabilities for first 5 test records (rounded to 3 decimals):")
print(pred_probabilities[:5].round(3))

# Show class names for probabilities
print(f"\nProbability classes order: {gnb.classes_}")

# Step 7: Calculate accuracy on the testing dataset
accuracy_test = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of Gaussian Naive Bayes on the test set: {accuracy_test:.4f} ({accuracy_test*100:.2f}%)")

# Additional metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Step 8: Predict class label for the first record of the testing dataset
first_record = X_test.iloc[:1]  # first row
first_pred_label = gnb.predict(first_record)
print("\nFirst test record features:")
print(first_record)
print(f"\nPredicted class label for the first test record: {first_pred_label[0]}")

# Step 9: Predict probability for the first record
first_pred_prob = gnb.predict_proba(first_record)
print("\nPredicted probabilities for the first test record:")
for class_name, prob in zip(gnb.classes_, first_pred_prob[0]):
    print(f"  {class_name}: {prob:.6f}")

# Step 10: Check if prediction for the first record is correct
first_actual_label = y_test.iloc[0]
first_accuracy = 1.0 if first_pred_label[0] == first_actual_label else 0.0
print(f"\nActual class label: {first_actual_label}")
print(f"Prediction correct: {first_pred_label[0] == first_actual_label}")
print(f"Accuracy on the first test record: {first_accuracy}")

# Bonus: Predict a new sample
print("\n" + "="*50)
print("BONUS: Predict a new iris flower")
sample = pd.DataFrame({
    'SepalLength': [5.1],
    'SepalWidth': [3.5],
    'PetalLength': [1.4],
    'PetalWidth': [0.2]
})
sample_pred = gnb.predict(sample)
sample_prob = gnb.predict_proba(sample)
print(f"Sample features: SepalLength=5.1, SepalWidth=3.5, PetalLength=1.4, PetalWidth=0.2")
print(f"Predicted class: {sample_pred[0]}")
print("Prediction probabilities:")
for class_name, prob in zip(gnb.classes_, sample_prob[0]):
    print(f"  {class_name}: {prob:.4f}")
```

```python
# 11. k-NN classifier using Euclidean distance - FIXED VERSION

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

# Step 1: Load dataset
data = pd.read_csv('dataset/iris.csv')

# Step 2: Assign proper column names
if len(data.columns) == 5:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
else:
    # If column names exist, use them
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']

# Step 2a: Data cleaning
print("Initial data shape:", data.shape)

# Replace empty strings or spaces with NaN
data = data.replace(r'^\s*$', np.nan, regex=True)

# Convert numeric columns to float
numeric_cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows with any NaN values
initial_rows = len(data)
data = data.dropna()
print(f"Rows after cleaning: {len(data)} (removed {initial_rows - len(data)} invalid rows)")

# Clean the Class column
data['Class'] = data['Class'].str.strip()

print(f"Unique classes: {data['Class'].unique()}")
print(f"Class distribution:\n{data['Class'].value_counts()}")

# Step 3: Split features and target
X = data[numeric_cols]  # Use only numeric columns
y = data['Class']

# Verify data types
print(f"\nFeature data types:\n{X.dtypes}")
print(f"\nFirst 5 rows of features:")
print(X.head())

# Step 4: Split dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nTraining set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")

# IMPORTANT: For k-NN, feature scaling is crucial
# Scale features to have mean=0 and variance=1 (standardization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nFeatures have been standardized for better k-NN performance")

# Step 5: Initialize k-NN classifier (k=3, Euclidean distance)
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(X_train_scaled, y_train)

# Step 6: Predict class labels for the test set
y_pred = knn.predict(X_test_scaled)

# Step 7: Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of k-NN classifier on Iris test set: {accuracy:.4f} ({accuracy*100:.2f}%)")

# Step 8: Predict class for first two test records
first_two_pred = knn.predict(X_test_scaled[:2])
print("\nFirst two test records (original features):")
print(X_test.iloc[:2])
print(f"\nPredicted classes for the first two test records: {first_two_pred}")
print(f"Actual classes for the first two test records: {y_test.iloc[:2].values}")

# Step 9: Confusion matrix
cm = confusion_matrix(y_test, y_pred)
target_names = sorted(y.unique())  # Dynamically get class names

# Create DataFrame with proper labels
conf_matrix_df = pd.DataFrame(cm, index=target_names, columns=target_names)
print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# Step 10: Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Bonus: Try different k values
print("\n" + "="*50)
print("BONUS: Comparing different k values")
print("="*50)

k_values = [1, 3, 5, 7, 9, 11, 13, 15]
accuracies = []

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn_temp.fit(X_train_scaled, y_train)
    y_pred_temp = knn_temp.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred_temp)
    accuracies.append(acc)
    print(f"k = {k:2d}: Accuracy = {acc:.4f} ({acc*100:.2f}%)")

# Find best k
best_k = k_values[np.argmax(accuracies)]
print(f"\nBest k value: {best_k} with accuracy {max(accuracies):.4f}")

# Step 11: Predict a new sample (with scaling)
print("\n" + "="*50)
print("PREDICTING NEW SAMPLES")
print("="*50)

# New sample features
new_sample = pd.DataFrame({
    'SepalLength': [5.1],
    'SepalWidth': [3.5],
    'PetalLength': [1.4],
    'PetalWidth': [0.2]
})

# Scale the new sample using the same scaler
new_sample_scaled = scaler.transform(new_sample)

# Predict
prediction = knn.predict(new_sample_scaled)
prediction_proba = knn.predict_proba(new_sample_scaled)

print(f"New sample features: {new_sample.iloc[0].to_dict()}")
print(f"Predicted class: {prediction[0]}")
print("\nPrediction probabilities:")
for class_name, prob in zip(knn.classes_, prediction_proba[0]):
    print(f"  {class_name}: {prob:.4f}")

# Step 12: Find k-nearest neighbors for the new sample
print("\n" + "="*50)
print("FINDING K-NEAREST NEIGHBORS")
print("="*50)

# Get distances and indices of k nearest neighbors
distances, indices = knn.kneighbors(new_sample_scaled, n_neighbors=3)

print(f"Distances to the 3 nearest neighbors: {distances[0]}")
print("Training samples of the 3 nearest neighbors:")
for i, idx in enumerate(indices[0]):
    print(f"  Neighbor {i+1}: {X_train.iloc[idx].to_dict()} -> Class: {y_train.iloc[idx]}")
```

```python
# 12. k-NN classifier using Manhattan distance and 7 neighbors - FIXED VERSION

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

# Load dataset
data = pd.read_csv('dataset/iris.csv')

# Assign column names
if len(data.columns) == 5:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
else:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']

# DATA CLEANING - Critical step
print("=== DATA CLEANING ===")
print(f"Original data shape: {data.shape}")

# Replace empty strings or spaces with NaN
data = data.replace(r'^\s*$', np.nan, regex=True)

# Convert numeric columns to float
numeric_cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows with any NaN values
rows_before = len(data)
data = data.dropna()
print(f"Removed {rows_before - len(data)} rows with invalid values")
print(f"Cleaned data shape: {data.shape}")

# Clean the Class column (strip whitespace)
data['Class'] = data['Class'].str.strip()

print(f"\nClasses found: {data['Class'].unique()}")
print(f"Class distribution:\n{data['Class'].value_counts()}")

# Features and target
X = data[numeric_cols]
y = data['Class']

# Verify data types and values
print(f"\nFeature data types:")
print(X.dtypes)
print(f"\nFirst 5 rows of features:")
print(X.head())
print(f"\nBasic statistics:")
print(X.describe())

# 70/30 train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nTraining set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")

# IMPORTANT: Scale features for k-NN (optional but recommended)
# For Manhattan distance, scaling is still important to give equal weight to all features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nFeatures have been standardized for better performance")

# k-NN classifier, 7 neighbors, Manhattan distance
knn = KNeighborsClassifier(n_neighbors=7, metric='manhattan')
knn.fit(X_train_scaled, y_train)

# Predict test set labels
y_pred = knn.predict(X_test_scaled)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of k-NN classifier (Manhattan, k=7): {accuracy:.4f} ({accuracy*100:.2f}%)")

# Confusion matrix without labels
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (without labels):")
print(cm)

# Confusion matrix with labels - DYNAMIC (no hardcoding)
target_names = sorted(y.unique())  # Get actual class names from data
conf_matrix_df = pd.DataFrame(cm, index=target_names, columns=target_names)
print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Optional: predict first two records
first_two_pred = knn.predict(X_test_scaled[:2])
print("\nFirst two test records (original features):")
print(X_test.iloc[:2])
print(f"\nPredicted classes for first two test records: {first_two_pred}")
print(f"Actual classes: {y_test.iloc[:2].values}")

# BONUS: Compare Euclidean vs Manhattan distance
print("\n" + "="*50)
print("COMPARISON: Manhattan vs Euclidean Distance")
print("="*50)

# Manhattan distance (k=7)
knn_manhattan = KNeighborsClassifier(n_neighbors=7, metric='manhattan')
knn_manhattan.fit(X_train_scaled, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
acc_manhattan = accuracy_score(y_test, y_pred_manhattan)

# Euclidean distance (k=7)
knn_euclidean = KNeighborsClassifier(n_neighbors=7, metric='euclidean')
knn_euclidean.fit(X_train_scaled, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
acc_euclidean = accuracy_score(y_test, y_pred_euclidean)

print(f"Manhattan distance accuracy: {acc_manhattan:.4f} ({acc_manhattan*100:.2f}%)")
print(f"Euclidean distance accuracy: {acc_euclidean:.4f} ({acc_euclidean*100:.2f}%)")
print(f"Difference: {abs(acc_manhattan - acc_euclidean):.4f}")

# BONUS: Try different k values with Manhattan distance
print("\n" + "="*50)
print("EXPLORING DIFFERENT K VALUES (Manhattan distance)")
print("="*50)

k_values = [1, 3, 5, 7, 9, 11, 13, 15]
manhattan_accuracies = []

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    knn_temp.fit(X_train_scaled, y_train)
    y_pred_temp = knn_temp.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred_temp)
    manhattan_accuracies.append(acc)
    print(f"k = {k:2d}: Accuracy = {acc:.4f} ({acc*100:.2f}%)")

best_k = k_values[np.argmax(manhattan_accuracies)]
best_acc = max(manhattan_accuracies)
print(f"\nBest k value for Manhattan distance: {best_k} with accuracy {best_acc:.4f} ({best_acc*100:.2f}%)")

# BONUS: Predict new sample
print("\n" + "="*50)
print("PREDICTING NEW SAMPLE")
print("="*50)

# Create a new sample
new_sample = pd.DataFrame({
    'SepalLength': [6.0],
    'SepalWidth': [2.7],
    'PetalLength': [5.1],
    'PetalWidth': [1.6]
})

# Scale the new sample
new_sample_scaled = scaler.transform(new_sample)

# Predict with Manhattan distance model
prediction = knn_manhattan.predict(new_sample_scaled)
prediction_proba = knn_manhattan.predict_proba(new_sample_scaled)

print(f"New sample: SepalLength=6.0, SepalWidth=2.7, PetalLength=5.1, PetalWidth=1.6")
print(f"Predicted class: {prediction[0]}")
print("\nPrediction probabilities:")
for class_name, prob in zip(knn_manhattan.classes_, prediction_proba[0]):
    print(f"  {class_name}: {prob:.4f}")

# Find the 7 nearest neighbors for this sample
distances, indices = knn_manhattan.kneighbors(new_sample_scaled, n_neighbors=7)
print(f"\nDistances to 7 nearest neighbors (Manhattan): {distances[0].round(4)}")
print("Classes of nearest neighbors:")
for i, idx in enumerate(indices[0]):
    print(f"  Neighbor {i+1}: {y_train.iloc[idx]}")
```

```python
# 13. Feature selection using Information Gain (Mutual Information) - FIXED VERSION

import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.preprocessing import LabelEncoder

# Load Iris dataset
data = pd.read_csv('dataset/iris.csv')

# Assign column names
if len(data.columns) == 5:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
else:
    data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']

# DATA CLEANING - Critical step
print("=== DATA CLEANING ===")
print(f"Original data shape: {data.shape}")

# Replace empty strings or spaces with NaN
data = data.replace(r'^\s*$', np.nan, regex=True)

# Convert numeric columns to float
numeric_cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows with any NaN values
rows_before = len(data)
data = data.dropna()
print(f"Removed {rows_before - len(data)} rows with invalid values")
print(f"Cleaned data shape: {data.shape}")

# Clean the Class column (strip whitespace)
data['Class'] = data['Class'].str.strip()

print(f"\nClasses found: {data['Class'].unique()}")
print(f"Class distribution:\n{data['Class'].value_counts()}")

# Separate features and target
X = data[numeric_cols]
y = data['Class']

# Verify data types and values
print(f"\nFeature data types:")
print(X.dtypes)
print(f"\nFirst 5 rows of features:")
print(X.head())
print(f"\nBasic statistics:")
print(X.describe())

# For mutual information, we need to encode the target variable if it's categorical
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"\nEncoded classes: {le.classes_}")
print(f"Class mapping: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# Step 1: Calculate Information Gain (Mutual Information) for each feature
mi_scores = mutual_info_classif(X, y_encoded, random_state=42)

# Step 2: Display scores
mi_series = pd.Series(mi_scores, index=X.columns)
print("\n" + "="*50)
print("Information Gain (Mutual Information) scores for each feature:")
print("="*50)
print(mi_series.sort_values(ascending=False))
print(f"\nTotal information gain: {mi_scores.sum():.4f}")

# Step 3: Select top 2 features based on Information Gain
k_best = SelectKBest(mutual_info_classif, k=2)
X_selected = k_best.fit_transform(X, y_encoded)

# Step 4: Print dataset with selected features
print("\n" + "="*50)
print("Dataset with top 2 selected features (first 10 rows):")
print("="*50)
# Create DataFrame with selected features for better visualization
selected_feature_indices = k_best.get_support(indices=True)
selected_feature_names = X.columns[selected_feature_indices]
X_selected_df = pd.DataFrame(X_selected, columns=selected_feature_names)
print(X_selected_df.head(10))

# Step 5: Shapes before and after feature selection
print("\n" + "="*50)
print("Data shapes comparison:")
print("="*50)
print(f"Shape of original data: {X.shape}")
print(f"Shape of data after selecting top 2 features: {X_selected.shape}")

# Step 6: Names of selected features
print(f"\nSelected features (k=2): {selected_feature_names.tolist()}")

# BONUS: Feature selection with different k values
print("\n" + "="*50)
print("BONUS: Feature selection with different k values")
print("="*50)

for k in [1, 2, 3, 4]:
    k_best_temp = SelectKBest(mutual_info_classif, k=k)
    X_selected_temp = k_best_temp.fit_transform(X, y_encoded)
    selected_indices = k_best_temp.get_support(indices=True)
    selected_names = X.columns[selected_indices]
    print(f"k = {k}: Selected features: {selected_names.tolist()}")

# BONUS: Feature importance visualization
print("\n" + "="*50)
print("BONUS: Feature importance analysis")
print("="*50)

# Sort features by information gain
sorted_features = mi_series.sort_values(ascending=False)

# Calculate percentage contribution
total_ig = mi_scores.sum()
feature_importance_percent = (sorted_features / total_ig * 100)

print("Feature importance (as percentage of total information gain):")
for feature, score in sorted_features.items():
    percent = (score / total_ig) * 100
    print(f"  {feature}: {score:.4f} ({percent:.1f}% of total)")

# BONUS: Create a simple model comparison
print("\n" + "="*50)
print("BONUS: Model comparison with and without feature selection")
print("="*50)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Model with all features
dt_full = DecisionTreeClassifier(random_state=42)
dt_full.fit(X_train, y_train)
y_pred_full = dt_full.predict(X_test)
acc_full = accuracy_score(y_test, y_pred_full)

# Model with top 2 features
X_train_selected = k_best.fit_transform(X_train, y_train)
X_test_selected = k_best.transform(X_test)
dt_selected = DecisionTreeClassifier(random_state=42)
dt_selected.fit(X_train_selected, y_train)
y_pred_selected = dt_selected.predict(X_test_selected)
acc_selected = accuracy_score(y_test, y_pred_selected)

print(f"Accuracy with all 4 features: {acc_full:.4f} ({acc_full*100:.2f}%)")
print(f"Accuracy with top 2 features: {acc_selected:.4f} ({acc_selected*100:.2f}%)")
print(f"Accuracy difference: {acc_full - acc_selected:.4f}")

if acc_selected >= acc_full:
    print("✓ Feature selection maintained or improved accuracy with fewer features!")
else:
    print("Note: Small decrease in accuracy but with 50% fewer features")

# BONUS: Visual comparison of feature importance
print("\n" + "="*50)
print("BONUS: Feature ranking summary")
print("="*50)

for rank, (feature, score) in enumerate(sorted_features.items(), 1):
    stars = "★" * min(5, int(score * 10)) if score > 0 else "☆"
    print(f"Rank #{rank}: {feature:12s} - IG = {score:.4f} {stars}")

# BONUS: Predict using only selected features
print("\n" + "="*50)
print("BONUS: Make predictions with selected features")
print("="*50)

# Create a new sample with all features
sample_full = pd.DataFrame({
    'SepalLength': [5.1],
    'SepalWidth': [3.5],
    'PetalLength': [1.4],
    'PetalWidth': [0.2]
})

# Extract only the selected features
sample_selected = sample_full[selected_feature_names]

# Predict using the model trained on selected features
sample_pred_encoded = dt_selected.predict(sample_selected)
sample_pred = le.inverse_transform(sample_pred_encoded)

print(f"Sample features: {sample_full.iloc[0].to_dict()}")
print(f"Selected features used: {selected_feature_names.tolist()}")
print(f"Predicted class: {sample_pred[0]}")
```
