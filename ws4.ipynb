```python


# 4.  Confusion Matrix, Accuracy, Precision, Recall, F1-Measure, ROC, AUC


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.metrics import roc_curve, auc, roc_auc_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder

data=pd.read_csv('dataset/predictions.csv')
actual=data['Actual']
predicted=data['Predicted']
# print (actual)
# print (predicted)

confusion_matrix = metrics.confusion_matrix(actual, predicted)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['Posetive', 'Negative'])
cm_display.plot()
plt.show()

precision = precision_score(actual, predicted, average='macro')
recall = recall_score(actual, predicted, average='macro')
f1 = f1_score(actual, predicted, average='macro')

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')



# Creating a instance of label Encoder.
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
data1 = le.fit_transform(data['Actual'])
data2 = le.fit_transform(data['Predicted'])

data3=[data1, data2]

fpr, tpr, thresholds = roc_curve(data1, data2)

print ("False Posetive rate = ", fpr)
print ("True Posetive Rate = ", tpr)
print ("The Threshold value = ", thresholds)

classes=['Posetive', 'Negative']
print(classification_report(actual, predicted, target_names=classes))

roc_auc = auc (fpr, tpr)
print(f'ROC AUC Score: {roc_auc:.2f}')

plt.figure(figsize=(4, 4))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

print(f"Calculated AUC: {roc_auc:.4f}")
```

```python
  
#7.	Decision tree algorithm using entropy (Information gain)


import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('dataset/iris.csv')

# Features and target
X = data.iloc[:,0:4]
Y = data.iloc[:,4]

# -------------------------------
# 80/20 Data Split
# -------------------------------
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, Y, test_size=0.2, random_state=100)

clf1 = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf1.fit(X_train1, y_train1)

print("Model trained with 80/20 split")

# -------------------------------
# 70/30 Data Split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X_train, y_train)

# -------------------------------
# Generate the Decision Tree
# -------------------------------
plt.figure(figsize=(12,8))
plot_tree(clf,
          feature_names=data.columns[0:4],
          class_names=clf.classes_,
          filled=True)
plt.title("Decision Tree (Entropy)")
plt.show()

# -------------------------------
# Tree in Text Format
# -------------------------------
tree_rules = export_text(clf, feature_names=list(data.columns[0:4]))
print("\nDecision Tree Rules:\n")
print(tree_rules)

# -------------------------------
# Prediction on Test Data
# -------------------------------
y_pred = clf.predict(X_test)

# -------------------------------
# Classification Report
# -------------------------------
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# -------------------------------
# Confusion Matrix (Without Label)
# -------------------------------
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (without class labels):\n")
print(conf_matrix)

# -------------------------------
# Confusion Matrix (With Labels)
# -------------------------------
labels = clf.classes_

conf_matrix_df = pd.DataFrame(conf_matrix,
                              index=labels,
                              columns=labels)

print("\nConfusion Matrix (with class labels):\n")
print(conf_matrix_df)

# -------------------------------
# Performance Metrics
# -------------------------------
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# -------------------------------
# Predict New Record
# -------------------------------
sample = [[6.4,1.8,6.6,2.1]]
prediction = clf.predict(sample)

print("\nClass label for record {6.4,1.8,6.6,2.1} is:", prediction)
```

```python
#8.	Decision tree algorithm using gini index (CART)
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('dataset/iris.csv')

# Features and target
X = data.iloc[:,0:4]
Y = data.iloc[:,4]

# Split dataset (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

# Decision Tree using Gini Index (CART)
clf = DecisionTreeClassifier(criterion="gini", random_state=100)

# Train model
clf.fit(X_train, y_train)

# Predict test data
y_pred = clf.predict(X_test)

# -------------------------------
# Classification Report
# -------------------------------
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# -------------------------------
# Confusion Matrix (without labels)
# -------------------------------
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (without labels):")
print(conf_matrix)

# -------------------------------
# Confusion Matrix (with labels)
# -------------------------------
labels = clf.classes_

conf_matrix_df = pd.DataFrame(conf_matrix,
                              index=labels,
                              columns=labels)

print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# -------------------------------
# Accuracy, Precision, Recall, F1
# -------------------------------
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# -------------------------------
# Predict new record
# -------------------------------
sample = [[6.4, 1.8, 6.6, 2.1]]
prediction = clf.predict(sample)

print("\nClass label for {6.4,1.8,6.6,2.1}:", prediction)

# -------------------------------
# Plot Decision Tree
# -------------------------------
plt.figure(figsize=(12,8))
plot_tree(clf,
          feature_names=data.columns[0:4],
          class_names=clf.classes_,
          filled=True)

plt.title("Decision Tree using Gini Index (CART)")
plt.show()

# -------------------------------
# Tree in Text Form
# -------------------------------
tree_rules = export_text(clf, feature_names=list(data.columns[0:4]))

print("\nDecision Tree Rules:\n")
print(tree_rules)
```

```python
# 9. Naïve Bayes algorithm on ‘Buys Computer’ dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score

# 1. Load Dataset
data=pd.read_csv('dataset/computer.csv')

print(data)
# 2. Encode nominal features into numerical representations
encoder = OrdinalEncoder()
data[['age', 'income', 'student', 'credit_rating', 'buys_computer']] = encoder.fit_transform(data[['age', 'income', 'student', 'credit_rating', 'buys_computer']])

print (data.head())

X_train = data[['age', 'income', 'student', 'credit_rating']]
y_train = data['buys_computer']



# 4. Initialize and train the Categorical Naive Bayes classifier
cnb = CategoricalNB()
cnb.fit(X_train, y_train)
X_test=X_train[:10]
y_test=y_train[:10]
# 5. Make predictions on the test set
y_pred = cnb.predict(X_test)

print("\nPredictions on test set:")
print(pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}))
# 6. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("\n \n Accuracy in %  :",  accuracy *100 )
```

```python
# 10. Gaussian Naive Bayes on Iris dataset

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Step 1: Load the dataset
# If your CSV has no header, we set header=None and define column names
data = pd.read_csv('dataset/iris.csv', header=None)

# Assign proper column names
data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']

# Step 2: Separate features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Step 3: Split dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 4: Initialize and train the Gaussian Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Step 5: Predict class labels on the testing dataset
y_pred = gnb.predict(X_test)
print("\nPredicted class labels for the test set:")
print(y_pred)

# Step 6: Predict probabilities of each class for the testing dataset
pred_probabilities = gnb.predict_proba(X_test)
print("\nPredicted probabilities for the test set (rounded to 3 decimals):")
print(pred_probabilities.round(3))

# Step 7: Calculate accuracy on the testing dataset
accuracy_test = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of Gaussian Naive Bayes on the test set: {accuracy_test:.2f}")

# Step 8: Predict class label for the first record of the testing dataset
first_record = X_test.iloc[:1]  # first row
first_pred_label = gnb.predict(first_record)
print("\nPredicted class label for the first test record:")
print(first_pred_label)

# Step 9: Predict probability for the first record
first_pred_prob = gnb.predict_proba(first_record)
print("\nPredicted probability for the first test record (rounded to 6 decimals):")
print(first_pred_prob.round(6))

# Step 10: Check if prediction for the first record is correct
first_actual_label = y_test.iloc[0]
first_accuracy = 1.0 if first_pred_label[0] == first_actual_label else 0.0
print(f"\nAccuracy on the first test record: {first_accuracy}")
```

```python
# 11. k-NN classifier using Euclidean distance
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Step 1: Load dataset
data = pd.read_csv('dataset/iris.csv', header=None)

# Step 2: Assign proper column names
data.columns = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Class']

# Step 3: Split features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Step 4: Split dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 5: Initialize k-NN classifier (k=3, Euclidean distance)
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(X_train, y_train)

# Step 6: Predict class labels for the test set
y_pred = knn.predict(X_test)

# Step 7: Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of k-NN classifier on Iris test set: {accuracy:.2f}")

# Step 8: Predict class for first two test records
first_two_pred = knn.predict(X_test.iloc[:2])
print("\nPredicted classes for the first two test records:")
print(first_two_pred)

# Step 9: Confusion matrix
cm = confusion_matrix(y_test, y_pred)
target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
conf_matrix_df = pd.DataFrame(cm, index=target_names, columns=target_names)
print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# Step 10: Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# 12. k-NN classifier using Manhattan distance and 7 neighbors
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Load dataset
data = pd.read_csv('dataset/iris.csv', header=None)
data.columns = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Class']

# Features and target
X = data.drop('Class', axis=1)
y = data['Class']

# 70/30 train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# k-NN classifier, 7 neighbors, Manhattan distance
knn = KNeighborsClassifier(n_neighbors=7, metric='manhattan')
knn.fit(X_train, y_train)

# Predict test set labels
y_pred = knn.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of k-NN classifier (Manhattan, k=7): {accuracy:.2f}")

# Confusion matrix without labels
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (without labels):")
print(cm)

# Confusion matrix with labels
target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
conf_matrix_df = pd.DataFrame(cm, index=target_names, columns=target_names)
print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))

# Optional: predict first two records
first_two_pred = knn.predict(X_test.iloc[:2])
print("\nPredicted classes for first two test records:")
print(first_two_pred)
```

```python
# 13. Feature selection using Information Gain (Mutual Information)

import pandas as pd
from sklearn.feature_selection import mutual_info_classif, SelectKBest

# Load Iris dataset
data = pd.read_csv('dataset/iris.csv', header=None)
data.columns = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Class']

# Separate features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Step 1: Calculate Information Gain (Mutual Information) for each feature
mi_scores = mutual_info_classif(X, y, random_state=42)

# Step 2: Display scores
mi_series = pd.Series(mi_scores, index=X.columns)
print("Information Gain (Mutual Information) scores for each feature:")
print(mi_series.sort_values(ascending=False))

# Step 3: Select top 2 features based on Information Gain
k_best = SelectKBest(mutual_info_classif, k=2)
X_selected = k_best.fit_transform(X, y)

# Step 4: Print dataset with selected features
print("\nDataset with top 2 selected features (first 5 rows):")
print(X_selected[:5])

# Step 5: Shapes before and after feature selection
print("\nShape of original data:", X.shape)
print("Shape of data after selecting top 2 features:", X_selected.shape)

# Step 6: Names of selected features
selected_feature_indices = k_best.get_support(indices=True)
selected_feature_names = X.columns[selected_feature_indices]
print("\nSelected features:", selected_feature_names.tolist())
```
