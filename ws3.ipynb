```python
# Solution on Data Preprocessing Exercise
# 1. Deliberately change the value of data to missing value in each feature of the dataset

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer

# Load the dataset
data11 = pd.read_csv('dataset/iris.csv')

iris_df = data11.iloc[:, :-1]

np.random.seed(42)  

for col in iris_df.columns:
    missing_indices = np.random.choice(iris_df.index, size=10, replace=False)
    iris_df.loc[missing_indices, col] = np.nan

print("DataFrame with missing values:")
print(iris_df)



```

```python

                    # 2.Impute the missing value data with mean value of the corresponding feature value

import pandas as pd

print("\nMissing values before imputation:")
print(iris_df.isnull().sum())

# Instead of using inplace=True
for col in iris_df.select_dtypes(include=['number']).columns:
    mean_value = iris_df[col].mean()
    iris_df[col] = iris_df[col].fillna(mean_value)   

print("\nDataFrame after mean imputation (numeric columns only):")
print(iris_df.head())
print("\nMissing values after imputation:")
print(iris_df.isnull().sum())

# Step 2: Assign Class column from data11
iris_df['Class'] = data11.iloc[:, 4]  # 5th column from data11

print("\nDataFrame after adding 'Class' column:")
print(iris_df.head())
```

```python
                      # 3.Adjust the precision to 2 round off. 

# code to round ofto 3 decimal place
iris_df=iris_df.round(3)
print(iris_df)
data=np.concatenate((iris_df,data11),axis=1)
print(data)
```

```python

# 4.Impute the missing value data with most frequent value of the corresponding feature

from sklearn.impute import SimpleImputer  

data2 = pd.read_csv('dataset/irisMissingValue.csv')

# Check last few rows and missing values
print(data2.tail())
print(round(data2.isnull().sum(), 2).sort_values(ascending=False))

# Impute missing values with most frequent value
imp_mode = SimpleImputer(strategy='most_frequent')  
data2_imputed = imp_mode.fit_transform(data2)       

# Convert back to DataFrame to keep column names
data2_imputed = pd.DataFrame(data2_imputed, columns=data2.columns)

print("\nData after imputing most frequent value:\n")
print(data2_imputed.tail())
```

```python
#             5.Impute the missing value data with a constant value 100 to every corresponding feature

#  Imputation with constant vales
data2 = pd.read_csv('dataset/irisMissingValue.csv')
print (data2.tail())
print (round(data2.isnull().sum(), 2).sort_values(ascending=False))
imputer = SimpleImputer(strategy='constant', fill_value=100)
imp_cons=imputer
imp_cons.fit(data2)
data2=imp_cons.transform(data2)
print (data2)
```

```python
               # 6.Impute the missing value data with KNN where N=2


data2 = pd.read_csv('dataset/irisMissingValue.csv')

# Remove class column (keep only features)
Before_imputation = data2.iloc[:, :-1]

# Show data before imputation
print("Data Before performing imputation:\n", Before_imputation)

# Create KNN Imputer (K = 2)
imputer = KNNImputer(n_neighbors=2)

# Apply imputation
# fit()
# Learns the structure of the dataset
# Computes distances between rows
# transform()
# Replaces missing values (NaN) using:
# → average of the 2 nearest neighbors
After_imputation = imputer.fit_transform(Before_imputation)

# Convert back to DataFrame
After_imputation = pd.DataFrame(
    After_imputation,
    columns=Before_imputation.columns
)

# Show result
print("\nAfter performing imputation:\n", After_imputation)
```
