```python
#1 Simple Linear regression
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error

data = pd.read_csv('dataset/ws5que1.csv')
print(data)
print("\nsize of the dataset", data.shape)
%matplotlib inline
plt.xlabel('Advertising (in Birr)')
plt.ylabel('Sales (in Birr)')
plt.scatter(data.Advertising,data.Sales,color='red',marker='*')
X = data.drop('Sales',axis='columns')
Y=data.Sales
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(data[['Advertising']],data.Sales)
print (reg.predict([[1.5]]))
print (reg.predict([[2]]))
print ("Slope = ", reg.coef_)
print ("X-intercep = ", reg.intercept_)
y=reg.coef_*1.5+reg.intercept_
print(y)
Y_calculated = reg.predict(X)
print ("\n\n Predicted Value \n ", Y_calculated)
print (Y_calculated.shape)
print ("\n\n Actual Value \n ", Y)
residual=Y-Y_calculated
print ("\n\n Residual (Error) \n", residual)
data['SalesCalculated']=reg.predict(X)
print (data)
data['residual'] = data.Sales - data.SalesCalculated
print (data)
print ("\n \n Mean Absolute Error (MAE): ", mean_absolute_error(data.Sales, data.SalesCalculated))
print ("\n \n Mean Squared Error (MSE): ", mean_squared_error(data.Sales, data.SalesCalculated))
print ("\n \n Root Mean Squared Error (RMSE):  ", root_mean_squared_error(data.Sales, data.SalesCalculated))
```

```python
# 3 Logistic regression on Gallstone dataset
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
data = pd.read_csv("D:\Downloads\gallstone-1\dataset-uci.xlsx ")
data.head()
print(data.shape)
X=data.drop('Gallstone Status', axis='columns')
Y=data['Gallstone Status']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,train_size=0.8)
# print (X_test.head())
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
model = LogisticRegression()
model.fit(X_train, y_train)

y_predicted = model.predict(X_test)
print("\n \n Calculated Value \n \n", y_predicted)
print("\n \n Actual Value \n ", y_test)
model.predict_proba(X_test)
conf_matrix = confusion_matrix(y_test, y_predicted)
print("Confusion Matrix:")
print(conf_matrix)
GallLabels=['Abscent', 'Present']
conf_matrix_df = pd.DataFrame(conf_matrix, index=GallLabels, columns=GallLabels)
print("\nConfusion Matrix (with labels):")
print(conf_matrix_df)
class_report = classification_report(y_test, y_predicted, target_names=GallLabels)
print("Classification Report:")
print(class_report)
print ("\n \n\ \n The slope of the dataset \n", model.coef_)
model.intercept_
temp1=model.coef_
temp2=model.intercept_
temp3=X.iloc[0,:]
temp4=temp1.dot(temp3) + temp2
print("\n \n temp4 value ", temp4)
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
yy=sigmoid(temp4)
print("\n \n sigmoid value ", yy)
```
