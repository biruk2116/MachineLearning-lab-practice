```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('dataset/WineQT.csv')

# 1. Dataset shape
print("\nShape:", data.shape)

# 2. First 5 rows
print("\nFirst 5 rows:\n", data.head())

# 3. Last 5 rows
print("\nLast 5 rows:\n", data.tail())

# 4. First 10 rows
print("\nFirst 10 rows:\n", data.head(10))

# 5. Statistical summary
print("\nDescription:\n", data.describe())

# 6. Class distribution (last column assumed label)
print("\nClass distribution:\n", data.iloc[:, -1].value_counts())

# 7. Features (X)
X = data.iloc[:, :-1].values
print("\nIndependent features (X):\n", X)

# 8. Labels (y)
y = data.iloc[:, -1].values
print("\nClass labels (y):\n", y)


# 📊 VISUALIZATION SECTION


# 9. Histogram
data.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()

# 10. Density plot (FIXED)
num_cols = data.shape[1]

cols = 3
rows = (num_cols // cols) + (num_cols % cols > 0)

data.plot(
    kind='density',
    subplots=True,
    layout=(rows, cols),
    sharex=False,
    figsize=(12, 10)
)

plt.tight_layout()
plt.show()

# 11. Boxplot (also fixed dynamically)
data.plot(
    kind='box',
    subplots=True,
    layout=(rows, cols),
    sharex=False,
    sharey=False,
    figsize=(12, 10)
)

plt.tight_layout()
plt.show()
```
